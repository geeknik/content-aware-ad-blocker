```python
class ASTSampler:
    def __init__(self, flags, flag_type="flag-any", count=5000):
        self.flags = flags # pandas df
        self._flag_type = flag_type
        self._count = count
        self._asts = [[], []]
        
    def sample(self, iter_ast):
        for id, ast in iter_ast:
            if not id in self.flags.index:
                continue
            flag_value = flags.loc[[id]][self._flag_type]
            if len(flag_value) > 1:
                continue
            else:
                flag_value = int(flag_value)
            asts = self._asts[flag_value]
            if len(asts) >= self._count:
                continue
            asts.append(json.loads(ast))
            if len(self._asts[0]) >= self._count and len(self._asts[1]) >=self._count:
                break
        l = min(len(self._asts[0]), len(self._asts[1]))
        self._asts[0] = self._asts[0][:l]
        self._asts[1] = self._asts[1][:l]
    @property
    def positive(self):
        return self._asts[1]
    @property
    def negative(self):
        return self._asts[0]
```

```python
sampler = ASTSampler(flags, count=500)
```

```python
def count_node(ast):
    count = 0
    if isinstance(ast, dict):
        if "type" in ast:
            count = count + 1
        for v in ast.itervalues():
            count = count + count_node(v)
    elif isinstance(ast, list):
        for v in ast:
            count = count + count_node(v)
    
    return count

def get_types_distinct(ast, types=set()):
    if isinstance(ast, dict):
        if "type" in ast:
            types.add(ast['type'])
        for v in ast.itervalues():
            get_types_distinct(v, types)
    elif isinstance(ast, list):
        for v in ast:
            get_types_distinct(v, types)
    return types

def get_types(ast, types=[]):
    if isinstance(ast, dict):
        if "type" in ast and ast['type'] != 'Program':
            types.append(ast['type'])
        for v in ast.itervalues():
            get_types(v, types)
    elif isinstance(ast, list):
        for v in ast:
            get_types(v, types)
    return types

def count_type(ast):
    if isinstance(ast, dict):
        if "type" in ast:
            yield ast['type'], 1
        for v in ast.itervalues():
            for x in count_type(v):
                yield x
    elif isinstance(ast, list):
        for v in ast:
            for x in count_type(v):
                yield x
```

```python
all_types = set()
for a in sampler.positive:
    get_types_distinct(a, all_types)
for a in sampler.negative:
    get_types_distinct(a, all_types)
```

```python
from collections import deque, Counter
```

```python
def strip_ast(ast):
    if isinstance(ast, dict) and 'type' in ast:
        tree = {}
        tree['node'] = ast['type']
        children = []
        for v in ast.itervalues():
            t = strip_ast(v)
            if t is not None:
                children.append(t)
            tree['children'] = deque(children)
        return tree
    elif isinstance(ast, list):
        return {
            "node" : 'ListNode',
            "children" : deque([strip_ast(v) for v in ast])
        }
    else:
        return None
```

```python
def binarize_node(tree):
    right = binarize_children(tree['children'])
    if right is None:
        return [tree['node'], None, right]
    else:
        return [tree['node'], right, None]
    
def binarize_children(children):
    if len(children) == 0:
        left = None
    elif len(children) == 1:
        left = children.popleft()
        left = binarize_node(left) 
    else:
        left = children.popleft()
        left = binarize_node(left)
        try:
            right = binarize_children(children)
        except:
            right = None
        if left[2] is None:
            left[2] = right
        else:
            left[1] = right
    return left
```

```python
def vectorize_btree(btree):
    counts = Counter()
    s = deque([btree])
    def get_name(x):
        if x is None:
            return ''
        elif isinstance(x, list):
            return x[0]
        else:
            return x
    while len(s):
        t = s.popleft()
        node, left, right = t
        key = "{}|{}|{}".format(
          get_name(node),
          get_name(left),
          get_name(right)
        )
        if left is not None:
            s.append(left)
        if right is not None:
            s.append(right)
            
        counts[key] += 1
    return counts
```

```python
def vectorize_ast(ast):
    tree = strip_ast(ast)
    btree = binarize_node(tree)
    return vectorize_btree(btree)
```

```python
features = [vectorize_ast(a) for a in all_asts]
```

```python
n = len(features)
sim_mat = np.zeros(dtype='float32', shape=[n, n])
```

```python
def similarity(c1, c2):
    keys = set(c1.keys()).union(c2.keys())
    s = 0.0
    for k in keys:
        diff = c1[k] - c2[k]
        s = s + diff*diff
    return s
```

```python
for i in xrange(n):
    for j in xrange(i, n):
        m = similarity(features[i], features[j])
        sim_mat[i, j] = m
        sim_mat[j, i] = m
```
